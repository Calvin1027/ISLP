```python
#Throughout this book, we will often want to generate the random data. Then, we can use np.random.normal() function
#there are three arguments of this functions; loc(mean), scale(standard deviation), and size(n).
x=np.random.normal(size=50) # 50 of random normal sample
x
y=x+np.random.normal(loc=50,scale=1, size=50)
#We can also calculate the correlation matrix between X and Y(상관계수)
np.corrcoef(x,y)

#In order to repeat the same result in our code, we can set a random seed-> using np.random.default_rng()
rng=np.random.default_rng(1303)
print(rng.normal(scale=5, size=2))
rng_2=np.random.default_rng(1303)
print(rng_2.normal(scale=5,size=2))
```

```python
#How to compute mean, variance, and standard deviation of arrays?
# we can just use np.mean(), np.var(), np.std()
rng=np.random.default_rng(3)
y=rng.standard_normal(10)
np.mean(y), y.mean()
```

```python
#For graphics, we download the library 'matplotlib'
from matplotlib.pyplot import subplots

fig,ax=subplots(figsize=(6,6)) #size of graphics that are shown in my screen
x=rng.standard_normal(100)
y=rng.standard_normal(100)
ax.plot(x,y);
```

```python
fig, axes=subplots(nrows=2,
                   ncols=3,
                   figsize=(15,5)) #make plot a form of matrix (rows=2, cols=3-> total 6)

axes[0,1].plot(x,y,'o')
axes[1,2].scatter(x,y,marker='+')
fig
```

```python
#We now create some more sophisticated plots-> ax.contour()
fig,ax=subplots(figsize=(8,8))
x=np.linspace(-np.pi,np.pi,50) #make 50 numbers from -pi to +pi
y=x #allocate same values to y
f=np.multiply.outer(np.cos(y),1/(1+x**2)) #outer is 외적 So it means outer product of cos(y) and (1+x**2)
ax.contour(x,y,f);

ax.contour(x,y,f,levels=45);
```

```python
#the ax.imshow() method can check the color depend on the z value like a heatmap
fig,ax=subplots(figsize=(8,8))
ax.imshow(f)
```

```python
#indexing Data
A=np.array(np.arange(16)).reshape((4,4))
A
```

```python
#To select multiple rows at a time
A[[1,3]]# rows
A[:,[0,2]] #columns

#suppose that we want to select the submatrix made up of the second and fourth rows as well as the first and third columns
A[[1,3]][:,[0,2]]

idx=np.ix_([1,3],[0,2,3]) #1,3행에서 0,2,3열의 원소 추출
A[idx]

A[1:4:2,0:3:2]
```

```python
#Boolean Indexing-> True or False?
keep_rows=np.zeros(A.shape[0],bool)
keep_rows
np.all(keep_rows==np.array([0,1,0,1])) #0=False, 1=True
```
